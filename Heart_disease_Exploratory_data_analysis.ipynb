## ***Heart Disease - Exploratory Data Analysis***


Heart disease, also known as cardiovascular disease, refers to a range of conditions that affect the heart. It's a major cause of illness and death globally, impacting millions of individuals each year. In an exploratory data analysis project focused on heart disease, researchers typically examine various datasets to uncover patterns, correlations, and insights related to this condition. The project might involve analyzing demographic factors (such as age, gender, and ethnicity), lifestyle choices (like diet and exercise habits), medical history (including pre-existing conditions like diabetes or hypertension), and specific cardiovascular measurements (such as blood pressure, cholesterol levels, and heart rate).

The goal of this analysis is to identify potential risk factors associated with heart disease and understand how these factors interact. Visualization techniques like histograms, scatter plots, and correlation matrices can be used to explore relationships within the data. Additionally, machine learning models may be employed to predict the likelihood of heart disease based on certain variables.

Overall, an exploratory data analysis project on heart disease aims to deepen our understanding of the factors influencing cardiovascular health, ultimately contributing to improved prevention and treatment strategies.

## Distribution Plot for Numeric Columns

For Numerical Columns we use bar charts, A bar chart is a graphical representation that uses rectangular bars to display data. Each bar represents a category or group, and the length of the bar corresponds to the value of the data it represents. Bar charts are effective for visualizing categorical data and comparing the frequency, count, or proportion of different categories.
In the context of exploratory data analysis (EDA) for heart disease, bar charts can be incredibly useful for several purposes:
1. Distribution of Categorical Variables
2. Comparision of Risk Factors
3. Visualizing Medical Conditions
4. Age Groups and Heart Disease

```python
sns.displot(df['Age'], kde = True)
```

As, we can see from the above graph, 50-60 years old person have high chances of heart disease.

```python
sns.distplot(df['Age'], kde = True, hist = True, rug = True)
```

```python
sns.distplot(df['Age'], bins = 30)
# bins make 30 bins in the graph
```

## Pie Charts for Categorical

A pie chart is a circular statistical graphic that is divided into slices to illustrate numerical proportions. Each slice of the pie represents a specific category, and the size of each slice is proportional to the quantity it represents relative to the whole.

In exploratory data analysis (EDA) for heart disease, pie charts can be valuable for visualizing categorical data and understanding the distribution or composition of different categories within a dataset. Here's how pie charts can be used effectively:
1. Distribution of Categorical Variables

## Violin Plot


A violin plot is a graphical representation that combines aspects of box plots and kernel density plots. It displays the distribution of continuous data across different categories or groups. The plot features a kernel density estimation on each side, mirrored around a central axis, resembling a violin's shape. The width of the "violin" at any given point represents the density of data points at that value.

In the context of exploratory data analysis (EDA) for heart disease, violin plots can be valuable for several purposes:
1. Comparision of Distributions
2. Identification of OUtliers
3. Visualization of Density
4. Exploring Relationships

```python
sns.violinplot(df['Age'])
```

```python
sns.violinplot(y = df['Sex'], x = df['HeartDisease'])
```

```python
sns.violinplot(y = df['Age'], x = df['HeartDisease'])
```

```python
sns.violinplot(y = df['Cholesterol'], x = df['HeartDisease'])
```

## Heatmap


A heat map is a graphical representation of data where values are depicted as colors within a matrix. It is particularly useful for visualizing the relationships and patterns in a dataset by displaying the magnitude of values as colors in a grid-like format.

In the context of exploratory data analysis (EDA) for heart disease, a heat map can be a powerful tool for several purposes:
1. Correlation Analysis
2. Risk Factor Identification
3. Comparision of Multivariate Data

```python
df1 = df.drop(columns=['Sex'])
df1 = df1.drop(columns=['ChestPainType'])
df1 = df1.drop(columns=['RestingECG'])
df1 = df1.drop(columns=['ExerciseAngina'])
df1 = df1.drop(columns=['ST_Slope'])
```

## Jointplot

A Jointplot is a type of statistical visualization that combines multiple plots to show the relationship between two variables along with their individual distributions. It is particularly useful for exploratory data analysis (EDA) to understand the correlation, distribution, and dependencies between variables.

In the context of heart disease EDA, a Jointplot can be valuable for several purposes:
1. Scatter Plot with Distributions
2. Correlation Analysis
3. Distribution Comaprision
4. Customization and Insights

```python
sns.jointplot(x = 'Age' , y = 'MaxHR', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df, kind ='reg')
```

```python
sns.jointplot(x = 'HeartDisease', y = 'Oldpeak', data = df)
```

```python
sns.jointplot(x = 'Cholesterol', y = 'Oldpeak', data = df)
```

```python
sns.jointplot(x = 'HeartDisease', y = 'MaxHR', data = df, kind = 'reg')
```

## Pairplot

A pairplot is a type of visualization in which pairwise relationships between multiple variables in a dataset are displayed in a grid of scatter plots and histograms. It allows for quick exploration of correlations and distributions across different combinations of variables, making it a powerful tool for exploratory data analysis (EDA).

In the context of heart disease EDA, a pairplot can be highly beneficial for several purposes:
1. Correlation Analysis
2. Distribution Visualization
3. Identification Of Patterns
4. Comparative Analysis
